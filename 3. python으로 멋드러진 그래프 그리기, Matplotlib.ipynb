## I. Matplotlib 시작하기

- 파이썬의 데이터 시각화 라이브러리
- `%matplotlib inline`을 통해서 활성화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## II. Matplotlib Case Study

```python
plt.plot([1, 2, 3, 4, 5]) # 실제 plotting 하는 함수
plt.show() # plt를 확인하는 명령
```

### Figsize : Figure(도면)의 크기 조절

```python
plt.figure(figsize=(10, 10)) # plotting을 할 도면을 선언
plt.plot([0, 1, 2, 3, 4]) # 실제 plotting 하는 함수
plt.show() # plt를 확인하는 명령
```

### 2차함수 그래프 with plot()

```python
# numpy.array를 이용해서 함수 그래프 그리기

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])

plt.plot(x, y)
plt.show()
```

```python
# np.arange(a, b, c) c: 0.01 

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

plt.plot(x, x**2)
plt.show()
```

```python
# x, y축의 범위를 설정하기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

###
plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]
###

plt.plot(x, x**2)
plt.show()
```

```python
# x, y축에 눈금 설정하기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

###
plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]
###

plt.xticks([i for i in range(-5, 6, 1)])
plt.yticks([i for i in range(0, 24, 3)])

plt.plot(x, x**2)
plt.show()
```

```python
# 그래프에 title 달기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]
plt.xticks([i for i in range(-5, 6, 1)])
plt.yticks([i for i in range(0, 24, 3)])

plt.title("y = x^2 graph")

# 범례 달기
plt.plot(x, x**2, label="trend")
plt.legend()

plt.show()
```

## III. Matplotlib Case Study

### 꺾은선 그래프(Plot)

```python
x = np.arange(20)
y= np.random.randint(0, 20, 20) # 난수를 20번 생성

plt.plot(x, y)
plt.show()

# Extra : y 축을 20까지 보이게 하고싶다면, y축을 5 단위로 보이게 하고 싶다면?
plt.axis([0, 20, 0, 20])
plt.yticks([i for i in range(0, 21, 5)])

plt.plot(x, y)
plt.show()
```

### 선점도(Scatter Plot)
- x와 y가 완전히 별개의 변수일 때 사용 -> 둘 사이의 상관 관계 파악

```python
plt.scatter(x, y)
plt.show()
```

### 박스 그림(Box plot)
- 수치형 데이터에 대한 정보 (Q1, Q2, Q3, min, max)

```python
plt.boxplot((x, y))

plt.title("Box plot of x, y")

plt.show()

```

### 막대 그래프(Bar Plot)
- 범주형 데이터의 "값"과 그 값의 크기를 직사각형으로 나타낸 그림

```python
plt.bar(x, y)

# plt.xticks([i for i in range(0, 20, 1)])
plt.xticks(np.arange(0, 20, 1))

plt.show()
```

```python
# cf) Histogram
# 도수분포를 직사각형의 막대 형태로 나타낸다.
# "계급"으로 그룹화하여 나타냄 "범주형" 데이터로 구성

plt.hist(y, bins= np.arange(0, 20, 2))

plt.xticks(np.arange(0, 20, 2))

plt.show()
```

### 원형 그래프(Pie Chart)
- 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그래프
- 다른 그래프에 비해서 **비율** 확인에 용이

```python
z = [100, 300, 200, 400]

plt.pie(z, labels=["one", "two", "three", "four"])
plt.show()
```

## IV. The 멋진 그래프, Seaborn Case Study

### Matplotlib를 기반으로 더 다양한 시각화 방법을 제공하는 라이브러리
- 커널밀도그림
- 카운트그림
- 캣그림
- 스트립그림
- 히트맵

```python
import seaborn as sns
```

### 커널밀도그림(Kernel Density Plot)
- 히스토그램과 같은 연속적인 분포를 곡선화해서 그린 그림
- `sns.kdeplot()`

```python
# in Histogram
x = np.arange(0, 22, 2)
y = np.random.randint(0, 20, 20)

plt.hist(y, bins=x)
plt.show()
```

```python
# kdeplot

sns.kdeplot(y, shade=True)
plt.show()
```

### 카운트그림(Count Plot)
- 범주형 column의 빈도수를 시각화 -> Groupby 후의 도수를 하는 것과 동일한 효과
- `sns.countplot()`

```python
vote_df = pd.DataFrame({"name" : ["Andy", "Bob", "Cat"], "vote":[True, True, False]})

vote_df
```

```python
# in matplotlib barplot

vote_count = vote_df.groupby("vote").count()

vote_count
```

```python
plt.bar(x=[False, True] , height=vote_count["name"])
plt.show()
```

```python
# sns의 countplot

sns.countplot(x=vote_df["vote"])
plt.show()
```

### 캣그림(Cat Plot)
- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수
- `sns.catplot()`

```python
covid = pd.read_csv("./country_wise_latest.csv")

covid.head(5)
```

```python
s = sns.catplot(x="WHO Region", y="Confirmed" , data=covid, kind="violin")
s.fig.set_size_inches(10, 6)
plt.show()
```

### 스트립그림(Strip Plot)
- scatter plot과 유사하게 데이터의 수치를 표현하는 그래프
- `sns.stripplot()`

```python
sns.stripplot(x="WHO Region" ,y="Recovered", data=covid)
plt.show()
```

```python
#cf) swarmplot
s = sns.swarmplot(x="WHO Region" ,y="Recovered", data=covid)

plt.show()
```

### 히트맵(Heatmap)
- 데이터의 행렬을 색상으로 표현하는 그래프
- `sns.heatmap()`

```python
covid.corr()
```

```python
sns.heatmap(covid.corr())

plt.show()
```
